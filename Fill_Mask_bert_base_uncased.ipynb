```python
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-base-uncased')
unmasker("Hello I'm a [MASK] model.")
```

**from transformers import pipeline**
 - Imports the Pipeline class from the Hugging Face Transformers package. The Pipeline class provides a simple interface for running NLP tasks without having to worry about low-level details like tokenization, batching, etc.

**unmasker = pipeline('fill-mask', model='bert-base-uncased')**
 - Creates a new instance of the Pipeline class specifically designed for filling masked tokens in sentences. It uses BERT base uncased model as its underlying architecture.

**unmasker("Hello I'm a [MASK] model.")**
 - Calls the unmasker object created earlier with a sentence containing a single masked word. This prompts the model to predict what the most likely word is that fits into the masked position based on context provided by the rest of the sentence.
